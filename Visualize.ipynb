```python
latent_space = torch.load('data/features.pt')
```

```python
latent_space_np = latent_space.numpy()
```

```python
print(np.max(latent_space_np))
```

```python
tsne = TSNE(n_components=2, learning_rate='auto', init='random')
latent_tsne = tsne.fit_transform(latent_space_np)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(latent_tsne[:, 0], latent_tsne[:, 1], cmap='viridis')
plt.title('t-SNE Visualization of Latent Space')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.show()
```
